```python
team_toss_split = {}
team_win_split = {}
```

```python
from collections import Counter
```

```python
def calculate_wins(x,team):
    if x.winner==team:
        if (x.toss_won==team and x.decision=='BOWL FIRST') or (x.toss_won!=team and x.decision=='BAT FIRST'):
            return 1
        elif (x.toss_won==team and x.decision=='BAT FIRST') or (x.toss_won!=team and x.decision=='BOWL FIRST'):
            return 0
    else:
        return 'lost'
```

```python
for team in df_season_summary_2022.home_team.unique():
    df_2022_team = df_season_summary_2022[(df_season_summary_2022.home_team==team)|(df_season_summary_2022.away_team==team)]
    bowl_first = dict(Counter(df_2022_team.apply(lambda x: 1 if ((x.toss_won==team and x.decision=='BOWL FIRST') or (x.toss_won!=team and x.decision=='BAT FIRST')) else 0,axis=1)))
    matches_won = dict(Counter(df_2022_team.apply(lambda x: calculate_wins(x,team),axis=1)))

    bowl_first['BOWL FIRST'],bowl_first['BAT FIRST'] = bowl_first[1],bowl_first[0]
    matches_won['BOWL FIRST'],matches_won['BAT FIRST'] = matches_won[1],matches_won[0] 
    del bowl_first[0],bowl_first[1],matches_won[0],matches_won[1],matches_won['lost']
    team_toss_split[team] = bowl_first
    team_win_split[team] = matches_won
    
```

```python
def plot_subplots(dictionary:dict,title_text:str,plot:str)->None:
    fig = make_subplots(rows=2,cols=5)
    row,col = 1,1
    
    for key,value in dictionary.items():
        if plot=='bar':
            labels = list(value.keys())
            values_toss_decision = list(value.values())
            fig.add_trace(go.Bar(x=labels,y=values_toss_decision,marker=dict(color=team_colors[key]),name=key),row=row,col=col)
         
        elif plot=='box':
            fig.add_trace(go.Box(y=value,name=key,marker=dict(color=team_colors[key])),row=row,col=col)
            
        col+=1
        if col>5:
            col = 1
            row = 2
    fig.update_layout(title_text=f'<b>{title_text}</b>',title_x=0.5)
    fig.show()
```

```python
bowlers_df = pd.DataFrame()
```

```python
def measure_bowler_performance(wickets_picked:int,economy:float,wickets_threshold:int,economy_threshold:float)->int:
    '''
    This fucntion is present solely for setting the size of the scatter plot for bowlers. The intention is to divide the graph into 4 quadrants and assign a marker 
    size which displays the value of their performance
    '''
    if wickets_picked>=wickets_threshold:
        if economy<=economy_threshold:
            return 40
        else:
            return 25
    else:
        if economy<=economy_threshold:
            return 30
        else:
            return 10
```

```python
def total_overs_bowled(df:pd.DataFrame)->int:
    '''
    This function returns the total overs bowled by a bowler during a specific phase of the game.
    '''
    total_overs = 0
    for match in df.match_id.unique():
        match_df = df[df.match_id==match] 
        total_overs += len(set(match_df.over))
    return total_overs
```

```python
def extract_bowlers_info(df:pd.DataFrame,bowler_name:str)->list:
    ''' 
     This function finds the following information for each phase:
        1) Wickets picked
        2) Over bowled
        3) Runs conceded
        4) Economy
    '''
    wickets_picked = np.shape(df[(df.wkt_bowler_name == bowler_name)&(df.wkt_text.str.contains("run out")==False)])[0]
    overs_bowled = total_overs_bowled(df)
    runs_conceded = df['runs'].sum()
    economy = runs_conceded/overs_bowled
    
    return [wickets_picked,overs_bowled,economy]
```

```python
def return_bowler_performance(df:pd.DataFrame,phase:str,bowler_name:str)->list:
    '''
    This function returns the collective data back to the base function
    '''
    if not df.shape[0]==0:
        bowler_data = extract_bowlers_info(df,bowler_name)
        return [bowler_data[0],bowler_data[1],bowler_data[2]]
    else:
        return[0,0,0]
```

```python
for bowler in df_season_details_2022.bowler1_name.unique():
    bowlers_info = {}
    bowlers_info['Bowler'] = bowler
    bowler_df = df_season_details_2022.loc[df_season_details_2022.bowler1_name==bowler,['match_id','over','runs','isWide','isNoball','bowler1_name','wkt_bowler_name','wkt_text']]

    wickets_picked = 0
    overs_bowled = 0
    economy = 0
    performance_rating = 0

    # Powerplay Analysis
    bowler_df_pp =  bowler_df[(bowler_df.over>=1) & (bowler_df.over<=6)]
    bowler_performance = return_bowler_performance(bowler_df_pp,'pp',bowler)
    
    wickets_picked,overs_bowled,economy = bowler_performance[0],bowler_performance[1],bowler_performance[2]
    
    if overs_bowled==0:
        performance_rating = 'NA'
    else:
        performance_rating = measure_bowler_performance(wickets_picked,economy,5,8.0)
    
    bowlers_info['Overs_in_PP'] = overs_bowled
    bowlers_info['Wickets_in_PP'] = wickets_picked
    bowlers_info['Economy_in_PP'] = economy
    bowlers_info['Performance_in_PP'] = performance_rating
    
    # Middle overs analysis
    bowler_df_middle =  bowler_df[(bowler_df.over>6) & (bowler_df.over<17)]
    bowler_performance = return_bowler_performance(bowler_df_middle,'middle',bowler)
    
    wickets_picked,overs_bowled,economy = bowler_performance[0],bowler_performance[1],bowler_performance[2]
    
    if overs_bowled==0:
        performance_rating = 'NA'
    else:
        performance_rating = measure_bowler_performance(wickets_picked,economy,5,8.0)
    
    bowlers_info['Overs_in_Middle'] = overs_bowled
    bowlers_info['Wickets_in_Middle'] = wickets_picked
    bowlers_info['Economy_in_Middle'] = economy
    bowlers_info['Performance_in_Middle'] = performance_rating

    # Death Overs Analysis
    bowler_df_death =  bowler_df[(bowler_df.over>=17)&(bowler_df.over<=20)]
    bowler_performance = return_bowler_performance(bowler_df_death,'death',bowler)
    
    wickets_picked,overs_bowled,economy = bowler_performance[0],bowler_performance[1],bowler_performance[2]
    
    if overs_bowled==0:
        performance_rating = 'NA'
    else:
        performance_rating = measure_bowler_performance(wickets_picked,economy,5,8.0)
                 
    bowlers_info['Overs_in_Death'] = overs_bowled
    bowlers_info['Wickets_in_Death'] = wickets_picked
    bowlers_info['Economy_in_Death'] = economy
    bowlers_info['Performance_in_Death'] = performance_rating
    
    bowlers_df = bowlers_df.append(bowlers_info,ignore_index=True)
```

```python
def plot_bowlers_performance(df:pd.DataFrame,xaxis_feature:str,yaxis_feature:str,marker_size_feature:str,marker_color_feature:str,title_text:str,wickets_threshold:int,economy_threshold:int)->None: 
    '''
    This function plots a scatter plot of the economy vs wickets taken for each bowler in a specific case. There are two dotted lines representing the 
    wickets threshold and economy threshold.
    '''
    fig = go.Figure(data=[go.Scatter(x=df[xaxis_feature],y=df[yaxis_feature],text=df.Bowler,mode='markers',name='Base Plot: Bowler Info',marker=dict(size=list(df[marker_size_feature]),color=df[marker_color_feature],colorscale='rainbow',showscale=True)),
                          go.Scatter(x=[x for x in range(25)],y=[wickets_threshold]*(25),mode='lines',line=dict(color='black',dash='dot',width=0.4),name='Wickets Threshold'),
                          go.Scatter(x=[economy_threshold]*(25),y=[y for y in range(25)],mode='lines',line=dict(color='black',dash='dot',width=0.4),name='Economy Threshold')])
    
    fig.update_layout(title_text=f'<b>Economy vs Wickets taken in {title_text}</b>',title_x = 0.5,xaxis=dict(tick0=min(df[xaxis_feature])),xaxis_title=f'<b>{xaxis_feature}</b>',
                      yaxis_title=f'<b>{yaxis_feature}</b>',legend=dict(yanchor="top",y=0.99,xanchor="left",x=0.01))
    fig.show()
```

```python
def measure_batsman_performance(runs_scored:float,strike_rate:float,runs_threshold:float,strike_rate_threshold:float)->int:
    '''
    Objective: To set the size of the scatter plot for batsmen based on the runs scored and the strike rate in each phase.
    '''
    if runs_scored>=runs_threshold:
        if strike_rate>=strike_rate_threshold:
            return 40
        else:
            return 25
    else:
        if strike_rate>=strike_rate_threshold:
            return 30
        else:
            return 10
```

```python
def plot_batsmen_performance(df:pd.DataFrame,xaxis_feature:str,yaxis_feature:str,marker_size_feature:str,marker_color_feature:str,title_text:str,xaxis_threshold:int,yaxis_threshold:int)->None: 
    '''
    Objective: To plot a scatter plot of the runs scored vs strike rate hit for each batsman in a specific case. There are two dotted lines representing the 
    thresholds for each axis.
    '''
    fig = go.Figure(data=[go.Scatter(x=df[xaxis_feature],y=df[yaxis_feature],text=df.Batsman,mode='markers',name='Base Plot: Batsman Info',marker=dict(size=list(df[marker_size_feature]),color=df[marker_color_feature],colorscale='rainbow',showscale=True)),
                          go.Scatter(x=[x for x in range(400)],y=[yaxis_threshold]*(400),mode='lines',line=dict(color='black',dash='dot',width=0.4),name='Runs threshold'),
                          go.Scatter(x=[xaxis_threshold]*(300),y=[y for y in range(300)],mode='lines',line=dict(color='black',dash='dot',width=0.4),name='Strike rate threshold')])
    
    fig.update_layout(title_text=f'<b>Runs vs Strike rate in {title_text}</b>',title_x = 0.5,xaxis=dict(tick0=min(df[xaxis_feature])),xaxis_title=f'<b>{xaxis_feature}</b>',
                      yaxis_title=f'<b>{yaxis_feature}</b>',legend=dict(yanchor="top",y=0.99,xanchor="left",x=0.01))
    fig.show()
```

```python
def calc_runs_scored(match:pd.DataFrame)->float:
    '''
    Objective: To return the runs scored by a batsman in a particular phase of the game.
    '''
    runs_scored = 0.0
    if float(match['batsman1_balls'].head(1))==1:
        runs_scored = float(match['batsman1_runs'].tail(1))
    else:
        if float(match['runs'].head(1))==0:
            runs_scored = float(match['batsman1_runs'].tail(1))-float(match['batsman1_runs'].head(1))
        else:
            runs = 0
            if match.isNoball.head(1).to_string()=='True':
                runs = float(match.runs.head(1))-1
            else:
                runs = float(match.runs.head(1))
            runs_scored = float(match['batsman1_runs'].tail(1))-float(match['batsman1_runs'].head(1))+runs
    return runs_scored
```

```python
def extract_batsmen_info(df:pd.DataFrame,phase:str)->list:
    '''
    Objective: To return a list containing the balls played, runs scored, times dismissed and boundaries hit for a batsman during a specific phase of the game.
    '''
    runs_scored = []
    times_out = 0
    balls_played = 0
    boundaries_hit = 0
    for id in df.match_id.unique():
        match = df[df.match_id==id]
        balls_played+= np.shape(match[(match.isWide==False)&(match.isNoball==False)])[0]
        runs_scored.append(calc_runs_scored(match))
        times_out+=len(match['wkt_batsman_name'].dropna())
        boundaries_scored = match['isBoundary'].astype(dtype=str)
        try: 
            boundaries = boundaries_scored.value_counts()['True']
            boundaries_hit += boundaries
        except:
            continue
    return [balls_played,np.sum(runs_scored),times_out,boundaries_hit]
```

```python
def return_batsman_performance(df:pd.DataFrame,phase:str)->list:
    '''
    Objective: Returns the data from extract_batsmen_info() to populate the batsmen_df DataFrame.
    '''
    if not df.shape[0]==0:
        batsman_data = extract_batsmen_info(df,phase)
        return [batsman_data[0],batsman_data[1],batsman_data[2],batsman_data[3]]
    else:
        return[0,0,0,0]
```

```python
for batsman in df_season_details_2022.batsman1_name.unique():
    batsman_info = {}
    batsman_info['Batsman'] = batsman
    batter_df = df_season_details_2022.loc[df_season_details_2022.batsman1_name == batsman,['match_id','over','runs','isWide','isNoball','isBoundary','batsman1_balls','batsman1_runs','wkt_batsman_name','wkt_batsman_runs']]
    
    balls_played = 0
    runs_scored = 0
    times_out = 0
    boundaries_hit = 0
    strike_rate = 0
    performance = 0
    
    # Powerplay analysis
    batter_df_pp = batter_df[(batter_df.over>=1) & (batter_df.over<=6)]
    balls_played,runs_scored,times_out,boundaries_hit = return_batsman_performance(batter_df_pp,'pp')
    if balls_played != 0:
        strike_rate = (runs_scored/balls_played)*100
        performance = measure_batsman_performance(runs_scored,strike_rate,160.0,120.0)
    
    batsman_info['Runs_scored_pp'] = runs_scored
    batsman_info['Balls_played_pp'] = balls_played
    batsman_info['Times_dismissed_pp'] = times_out
    batsman_info['Strike_rate_pp'] = strike_rate
    batsman_info['Boundaries_pp'] = boundaries_hit
    batsman_info['Performance_pp'] = performance
    
    # Middle overs analysis
    batter_df_middle = batter_df[(batter_df.over>6) & (batter_df.over<17)]
    performance = 0
    balls_played,runs_scored,times_out,boundaries_hit = return_batsman_performance(batter_df_middle,'middle')
    if balls_played != 0:
        strike_rate = (runs_scored/balls_played)*100
        performance = measure_batsman_performance(runs_scored,strike_rate,200.0,130.0)

    batsman_info['Runs_scored_middle'] = runs_scored
    batsman_info['Balls_played_middle'] = balls_played
    batsman_info['Times_dismissed_middle'] = times_out
    batsman_info['Strike_rate_middle'] = strike_rate
    batsman_info['Boundaries_middle'] = boundaries_hit
    batsman_info['Performance_middle'] = performance
    
    # Death overs analysis
    batter_df_death = batter_df[(batter_df.over>=17) & (batter_df.over<=20)]
    performance = 0
    balls_played,runs_scored,times_out,boundaries_hit = return_batsman_performance(batter_df_death,'death')
    if balls_played != 0:
        strike_rate = (runs_scored/balls_played)*100
        performance = measure_batsman_performance(runs_scored,strike_rate,140.0,160.0)

    batsman_info['Runs_scored_death'] = runs_scored
    batsman_info['Balls_played_death'] = balls_played
    batsman_info['Times_dismissed_death'] = times_out
    batsman_info['Strike_rate_death'] = strike_rate
    batsman_info['Boundaries_death'] = boundaries_hit
    batsman_info['Performance_death'] = performance
    
    batsmen_df = batsmen_df.append(batsman_info,ignore_index=True)
```
